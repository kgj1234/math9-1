# Tracking Variable Names and Values

```python
x=1
y=2
print(x,y)
y=x
x=y
print(x,y)
```

```python
x=1
y=2
print(x,y)
y=x
x=3
print(x,y)
```

What if we want to swap x and y. Two ways:

```python
x=1
y=2
print(x,y)
z=x
x=y
y=z
print(x,y)
```

```python
# Or we can switch the variables simultaneously; Not all languages can do this
x=1
y=2
print(x,y)
x,y=y,x
print(x,y)
```

# Problem

a) Plot the functions $f(x)=x^2\cos(x)+2x+1$ and $g(x)=x^3+2x^2+3x$ in the same plot, labelling the plot using a legend, giving it a title, and labeling the axes

```python
x=np.linspace(-10,10,2000)
f= lambda x: x**2*np.cos(x)
g=lambda x: x**3+2*x**2+3*x
plt.plot(x,f(x),label='function 1')
plt.plot(x,g(x),label='function 2')
plt.legend()
plt.title('problem 1')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

# If-Else Statements

```python
def abs(x):
    if x > 0:  # this is a comparison (more on this later)
        return x
    else:
        return -x
math9.graphfunction(abs)
```

# Boolean Variables

There is a variable type called bool. It takes values of either true or false. When testing conditions, the value obtained is a boolean, that tells us the truth value of the statement.

```python
print(1==1)
```
